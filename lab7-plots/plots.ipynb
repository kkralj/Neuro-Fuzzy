```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def neuron(x, w, s):
    return 1. / (1 + abs(x - w) / abs(s))
```

```python
X = np.linspace(-8, 10, 100)

for s in [0.25, 1, 4]:
    y = neuron(X, 2, s)
    plt.plot(X, y, label=str(s))

plt.legend()
plt.show()
```

```python
def readPoints(path):
    X1, X2, Y = [], [], []
    with open(path, 'r') as file:
        lines = file.read().splitlines()
        for line in lines:
            x1, x2, y1, y2, y3 = line.split()
            X1.append(float(x1))
            X2.append(float(x2))
            Y.append(np.argmax([y1, y2, y3]))
    return np.array(X1), np.array(X2), np.array(Y)
```

```python
X1, X2, Y = readPoints('../lab7-data/zad7-dataset.txt')
markers = ['s', 'o', '^']
colors = ['r', 'y', 'b']

for i in range(len(X1)):
    plt.scatter(X1[i], X2[i], marker=markers[Y[i]], color=colors[Y[i]])
plt.show()
```

```python
def readNeuronPoints(path):
    points = []
    with open(path, 'r') as file:
        lines = file.read().splitlines()
        for line in lines:
            points.append(float(line))
    return int(points[0]), points[1:]
```

```python
first_hidden, points = readNeuronPoints('../lab7-data/train-params-283.txt')
Uxi, Uyi = [], []
for i in range(0, first_hidden * 4, 4):
    Uxi.append(points[i])
    Uyi.append(points[i + 2])
    
for i in range(len(X1)):
    plt.scatter(X1[i], X2[i], marker=markers[Y[i]], color=colors[Y[i]])
plt.scatter(Uxi, Uyi, c='g')
    
plt.show()
```

```python
print('W')
for i in range(len(Uxi)):
    print(points[4 * i], points[4 * i + 2])

print('\nS')
for i in range(len(Uxi)):
    print(points[4 * i + 1], points[4 * i + 3])
    
print('\nIzlazni sloj')
for i in range(first_hidden * 4, len(points)):
    if (i - first_hidden * 4) % 9 == 0 and (i - first_hidden * 4) > 0:
        print()
    print(points[i])
```
