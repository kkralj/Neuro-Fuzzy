```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import normalize
```

```python
def read2DPoints(path):
    x, y = [], []
    with open(path, 'r') as file:
        lines = file.read().splitlines()
        for line in lines:
            x1, y1 = line.split()
            x.append(float(x1))
            y.append(float(y1))
    return np.array(x), np.array(y)

def normalize(X):
    return (X - min(X)) / (max(X) - min(X))
```

```python
# Membership functions plot
plt.subplots(figsize=(10, 45))

for i in range(1, 6 + 1):
    x, y = read2DPoints('../lab6-data/memberships/rule-' + str(i) + '-membership-A.dat')
    plt.subplot(10, 2, 2 * i - 1)
    plt.plot(normalize(x), y)
    plt.title('rule-' + str(i) + '-membership-A.dat')
    
    x, y = read2DPoints('../lab6-data/memberships/rule-' + str(i) + '-membership-B.dat')
    plt.subplot(10, 2, 2 * i)
    plt.plot(normalize(x), y)
    plt.title('rule-' + str(i) + '-membership-B.dat')
    
plt.show()
```

```python
from mpl_toolkits.mplot3d import axes3d, Axes3D 
```

```python
def read3DPoints(path):
    xs, ys, zs = [], [], []
    with open(path, 'r') as file:
        lines = file.read().splitlines()
        for line in lines:
            x, y, z = line.split()
            xs.append(float(x))
            ys.append(float(y))
            zs.append(float(z))
    return np.array(xs), np.array(ys), np.array(zs)
```

```python
# Output error plot
xs, ys, zs = read3DPoints('../lab6-data/errors/output-error.dat')

print('Max error:', max(zs))
print('Avg error:', np.mean(zs))

fig = plt.figure(1, figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xs, ys, zs)

plt.show()
```

```python
# Plot batch and stohastic error
axes = plt.gca()
axes.set_ylim([0, 1])
ind, errors = read2DPoints('../lab6-data/errors/batch-errors.dat')
plt.plot(ind, errors, label='batch')
plt.legend()
plt.show()

axes = plt.gca()
axes.set_ylim([0, 1])
ind, errors = read2DPoints('../lab6-data/errors/stohastic-errors.dat')
plt.plot(ind, errors, label='stohastic')
plt.legend()
plt.show()
```

```python
# Plot learning rate errors
plt.figure(1)
axes = plt.gca()
axes.set_ylim([0, 1000])
plt.title('Batch')

for lr in ['0.01', '0.001', '1.0E-4']:
    x, y = read2DPoints('../lab6-data/learningRates/batch-errors-' + lr + '.dat')
    plt.plot(x, y, label=lr)
    
plt.legend()
plt.show()

plt.figure(2)
plt.title('Stohastic')
axes = plt.gca()
axes.set_ylim([0, 5000])

for lr in ['0.1', '0.001', '1.0E-4']:
    x, y = read2DPoints('../lab6-data/learningRates/stohastic-errors-' + lr + '.dat')
    plt.plot(x, y, label=lr)
    
plt.legend()
plt.show()
```
